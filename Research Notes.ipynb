#### 1. Model

In <font color="#4682B4"> S.Mukherjee et al. 2021 </font> work, there are three essential parameters to model the scattering:

* **initial distance $r_{i}$** is taken to be within the radius of the cluster $r_{i} \le R_{c}$

* **initial velocity $v_{i}$** is obtained by employing the virial theorem : $v_{i} = \sqrt{\frac{G \overline{M}}{3 \overline{R}}} = \sqrt{\frac{G n_\mathrm{star} m}{3 R_{c}}}$

* **initial angle $\theta_{i}$** is the angle between the binary components, <u>as well as the angle between the initial velocity $v_{i}$ and initial distance $r_{i}$,

    can be arbitrary. The minimum $\theta_{i}$ is evaluated by assuming that periastron distance $r_{p}$ is always greater than twice of the schwarzschild radius $r_{s}$; the maximum $\theta_{i}$ is obtained by assuming that the SNR has to exceed threshold.

    Note that this <u>$\theta_{i}$ lies in the same orbital plane with $\phi$ and is not the $\theta$ from the spherical coordinate system.

#### 2. Derivation of the equation of orbit

Invoking the conservation of energy and angular momentum. Within the polar coordinates, the velocity of mass $m_{2}$ can be parameterized as 

$$ 
\begin{aligned}
\mathbf{v} &= v_{r}\hat{r} + v_{\phi}\hat{\phi} \\
& = \frac{dr}{dt}\hat{r} + r\frac{d \phi}{dt}\hat{\phi}
\end{aligned}
$$

The energy radiated in GW per unit mass is
$$ 
\begin{aligned}
E &= E_{k}+\Phi(r) \\
& = \frac{1}{2}\bigg[\bigg(\frac{dr}{dt}\bigg)^{2}+ r^{2}\bigg(\frac{d \phi}{dt}\bigg)^{2} \bigg]+ \Phi(r)
\end{aligned}
$$

According to the definition of angular momentum: $L:= \mathbf{r}\times \mathbf{v} = r^{2}\frac{d\phi}{dt} = r_{i}v_{i}\sin{\theta_{i}}$ and using the substitution $u = \frac{1}{r}$, we can obtain

$$
\frac{2E}{L^{2}} = \bigg(\frac{du}{d \phi}\bigg)^{2}+u^{2}+\frac{2\Phi(r)}{L^{2}} = const.
$$

Next, by differentiating both sides of the above equation, we can transform the equation into the following form due to constant term $\frac{2E}{L^{2}}$:

$$
0 = 2\frac{d^{2} u }{d \phi^{2}}+2u+\frac{2}{L^{2}}\frac{d \Phi(r)}{d u}
$$

After simplifying this equation, we get:

$$
\frac{d^{2} u }{d \phi^{2}}+u=-\frac{1}{L^{2}}\frac{d \Phi(r)}{d u}
$$

Given that $\frac{d \Phi(r)}{dr}= F(r) = -\frac{GM}{r^{2}}$, and using the substitution $dr = -\frac{1}{u^{2}} du$, we can get $-u^{2}\frac{d \Phi(u)}{du} = F(\frac{1}{u}) = -GMu^{2}$. 

This indicates that the equation is an inhomogeneous linear differential equation of second order, 

$$
\frac{d^{2} u }{d \phi^{2}}+u=-\frac{GM}{L^{2}}
$$

And the corresponding solution is 

$$
u(\phi) = \frac{1}{r(\phi)} = B \cos(\phi-\phi_{0})+\frac{GM}{L^{2}}
$$
$$
r(\phi) = \frac{L^{2}/GM}{1+(L^{2}/GM r_{p}-1))\cos(\phi-\phi_{0})}
$$

Due to the geometric properties of hyperbolic orbits $r= \frac{p}{1+e\cos(\phi-\phi_{0})}$, we can infer the eccentricity $e$ and semi-latus retum $p$ as:

$$
p = \frac{GM}{L^{2}}, \text{ and } e= \frac{L^{2}}{GMr_{p}}-1
$$


Based on the initial conditions and the orbital equation, we can easily obtain the 

$$
e^{2} = 1+... 
$$

$$
r_{p} = ... = \frac{L^{2}}{GM(1+e)}
$$

$$

$$

### 4. GWs from Hyperbolic Interactions

------
> <font color="#808080">  This section of notes is compiled from these three papers: <font color="#4682B4">Capozziello & De Laurentis 2008 </font> & <font color="#4682B4">De Vittori et al. 2012 </font> & <font color="#4682B4">Garca-Bellido & Nesseris 2018 </font>. Subsequent papers that calculate the quadrupole method have made minor modifications to these three foundational articles, such as changes in reference frames and initial conditions. However, the underlying method for their calculations remains consistent.</font>



<br> GWs are solutions of the linearized field equations of GR and the power radiated power to the leading order is given by Einstein's quadrupole formula:

$$
P = -\frac{G}{45 c^{5}} \langle \overset{...} D_{ij}\, \overset{...}D^{ij}\rangle
$$

where
$\dddot{D}_{ij} \dddot{D}^{ij} = \sum_{i,j} \dddot{D}_{ij} \dddot{D}^{ij}$
can be simplified as

$$
\left\langle \dddot{D}_{ij} \dddot{D}^{ij} \right\rangle = 6 \left\langle \dddot{M}_{11}^2 + \dddot{M}_{22}^2 + 3 \dddot{M}_{12}^2 - \dddot{M}_{11} \dddot{M}_{22} \right\rangle.
$$


It is convenient to rewrite the Cartesian coordinates $x_{i}$ in spherical coordinates

$$
\begin{aligned}
x &= r \cos \phi \sin \theta = r \cos \phi, \\
y &= r \sin \phi \sin \theta = r \sin \phi, \\
z &= r \cos \theta = 0
\end{aligned}
$$

where $\theta = \frac{\pi}{2}$. Thus, we obtain the components of the quadrupole mass tensor:

$$
\begin{aligned}
D_{11} &= \mu r^2 (3 \cos^2 \phi - 1), \\
D_{22} &= \mu r^2 (3 \sin^2 \phi - 1), \\
D_{12} &= D_{21} = 3 \mu r^2 \cos \phi \sin \phi, \\
D_{33} &= -\mu r^2,
\end{aligned}
$$

which has the following matrix form:

$$
D_{ij} = \mu r^2 (\phi) \begin{pmatrix}
3 \cos^2 \phi - 1 & 3 \cos \phi \sin \phi & 0 \\
3 \cos \phi \sin \phi & 3 \sin^2 \phi - 1 & 0 \\
0 & 0 & -1
\end{pmatrix}
$$



#### 1. The radiated power and total energy

The total radiated energy of this system can be determined knowing the impact parameter $b$, initial velocity $v_{i}$ and reduced mass $\mu$.

$$
\Delta E = \displaystyle \int_{-\infty}^{\infty} \frac{dE}{dt} dt = \displaystyle \int_{-\infty}^{\infty} P(t) dt
$$

$$

$$
f(\phi, \phi_0) = \frac{\sin^4(\phi_0 - \frac{\phi}{2}) \sin^4(\frac{\phi}{2}) }{\tan^{2} \phi_0 \sin^{6} \phi_0 }\\
\left\{ 150 + 72 \cos 2\phi_0 + 66 \cos 2(\phi_0 - \phi) - 144 \cos (2\phi_0 - \phi) - 144 \cos \phi \right\}
$$

$$

$$

$$
P(t) = \frac{dE}{dt} = -\frac{32G \mu^2 \alpha^4 v_i^6}{45c^5 b^2} f(\phi, e),

### 5. Encounter Event Rates
-------
They calculate the rate for hyperbolic encounters as a function of initial conditions $r_{i}$, $\theta_{i}$.

The steps are:
1. calculate the probability of fraction $P_{\theta_{i}}$ within the solid angle $\Omega(\theta_\mathrm{max}(r_{i}),\theta_\mathrm{min}(r_{i}))$;

3. calculate the event rate per unit time for an individual object inside the cluster $P_\mathrm{indv}$;

   
5. calculate the event rate throughout the entire cluster $P_\mathrm{clus}$;

   
7. calculate the event rate throughout the entire galaxy $n_\mathrm{gc}P_\mathrm{clus}$;

   
9. calculate the total event rate between redshift $z$ to $z+dz$.


#### 1. Methods
According to <font color="#4682B4">S.Mukherjee et al. 2021:

The total event rate is given as $$P_\mathrm{tot} = \displaystyle \int_{z_\mathrm{max}}^{z_\mathrm{min}} (1+z)^{-1} n_\mathrm{gc}P_\mathrm{clus} \mathcal{N}(z) dz $$

where,
* $n_\mathrm{gc}$ is the number of GCs in one galaxy, so the $n_\mathrm{gc} P_\mathrm{clus}$ probably means that the event rate $P_\mathrm{gal}$ of all the GCs in one galaxy.
* At a comoving distance $r$, the number of galaxies between redshift $(z, z+dz)$ is given as $\mathcal{N}(z) = \frac{d \mathcal{N}}{dz} = \frac{4\pi r^{2}}{H_{0}}\frac{cn(z)}{E(z)}$. $H_{0} = 67.4 \;\mathrm{km/s/Mpc}$ is the Hubble constant; $E(z) = \sqrt{\Omega_{\Lambda}+(1+z)^{3}\Omega_{m}}$, $\Omega_{m} = 0.315$, $\Omega_{\Lambda} = 0.685$. (Data from <font color="#4682B4">Aghanim et al. 2020</font>)
* the number density of MWEG is $n(z)$ which can be assumed a constant here

* $P_\mathrm{clus} = n_\mathrm{star}P_\mathrm{indv}$

where,
* $n_\mathrm{star}$ is the total number of stars


The individual probability:
$$
\begin{aligned}
P_\mathrm{indv} &= \frac{3 v_{i} n_\mathrm{star} \bigg(b^{2}_\mathrm{max}-b^{2}_\mathrm{min} \bigg)}{4R^{3}_{c}}\ln(R_{c}/R_\mathrm{min}) \\
 &= n_{s}v_{i}\pi\bigg(b^{2}_\mathrm{max}-b^{2}_\mathrm{min} \bigg) \ln(R_{c}/R_\mathrm{min}) \\
\end{aligned}
$$

where,
* $b_\mathrm{min} = r_{i}\theta_\mathrm{min}(r_{i})$
* $R_\mathrm{min}$ is the lowest radial cut-off, which can be obtained from this expression : $\frac{4\pi}{3}R_\mathrm{min}^{3}n_\mathrm{co} \sim \frac{4\pi}{3} R_\mathrm{co}^{3}$. This indicates the fact that the total volume of the cluster is composed of $n_\mathrm{co}$ number of compact objects with radius $R_\mathrm{star}$ distributed within radius $R_\mathrm{co}$

<br>

Using the approximation $\cos \theta \sim 1-\frac{\theta^{2}}{2}$, the probability of selecting this fraction of particles falls within the above solid angle is 
$$
P_{\theta_{i}} = \frac{\Omega}{4\pi} = \frac{1}{4\pi}\bigg[2\pi \displaystyle \int_{\theta_\mathrm{max}(r_{i})}^{\theta_\mathrm{min}(r_{i})}  \sin \theta_{i} d\theta_{i}\bigg] = \frac{1}{4}\bigg[\theta_\mathrm{max}^{2}(r_{i})-\theta_\mathrm{min}^{2}(r_{i})\bigg]
$$

For an individual object inside the cluster, the event rate or number of events per unit time becomes:

$$
P_\mathrm{indv} = \displaystyle \int_{R_\mathrm{min}}^{R_{c}}\bigg(\frac{P_{\theta}}{t_\mathrm{col}}\bigg)4\pi r_{i}^{2}n_{s}dr_{i}
$$


<br>
Note: All $\sin \theta$ were approximated to $\theta$, since $\theta$ given from the data are observed very small.

scale factor: $a = (1+z)^{-1}$

$$
D_{c}(z) = D_{H}\int_{0}^{z} \frac{d z^{\prime}}{E(z^{\prime})}
$$

The Hubble distance:

$$
D_{H} = \frac{c}{H_{0}} \approx 3000 \, h^{-1} \, \mathrm{Mpc}
$$

where the $H_{0}$ is the Hubble parameter today, $h$ is the dimensionless Hubble constant.

$$
E(z) = \frac{H(z)}{H_{0}}=\sqrt{\Omega_{r}(1+z)^{4}+\Omega_{m}(1+z)^{3}+\Omega_{k}(1+z)^{2}+\Omega_{\Lambda}}
$$

the present normalized values of the cosmological constant:

* $\Omega_{r}$: radiation energy density
  
* $\Omega_{m}$: matter density

  
* $\Omega_{\Lambda}$: dark energy density


* $\Omega_{k} = 1- \Omega_{r}-\Omega_{m}-\Omega_{\Lambda}$: curvature

So the dimensionless Hubble parameter at a given redshift is $H(z) = H_{0}E(z)$.


<br/>

then $E(z) = \sqrt{\Omega_{\Lambda}+ \Omega_{m}(1+z)^{3}}$. <font color="#4682B4"> Mazumder et al. 2014</font>

$\Omega_{m} = 1, \Omega_{k} = 0, \Omega_{\Lambda} = 0$, so 

$$
H(z) = H_{0}\cdot (1+z)^{3/2}
$$

In today's consensus cosmology: $\Omega_{m} = 0.315, \Omega_{k} = 0, \Omega_{\Lambda} = 0.689$, then 

$$
H(z) = H_{0}\sqrt{\Omega_{\Lambda}+ \Omega_{m}(1+z)^{3}}
$$
